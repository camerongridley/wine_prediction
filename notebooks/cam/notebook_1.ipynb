```python
y = df['quality']
X = df.drop(['quality'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
def run_model(model):
    model.fit(X_train_scaled, y_train)
    y_hat = model.predict(X_test_scaled)
    print(f'R2: {model.score(X_test_scaled, y_test)}')
    print(f'MAE: {mean_absolute_error(y_test, y_hat)}')
    plt.scatter(y_test, y_hat)
```

```python
lin_reg = LinearRegression()
run_model(lin_reg)
```

```python
lasso = Lasso(alpha=.0000001, random_state=42)
run_model(lasso)
```

```python
ridge = Ridge(alpha =.01, random_state=42)
run_model(ridge)
```

```python
enet = ElasticNet(alpha=.00001, random_state=42)
run_model(enet)
```

```python
rfr = RandomForestRegressor(random_state=42)
run_model(rfr)
```
